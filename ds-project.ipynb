```python
ethnicity_map = {
    0: 'Caucasian',
    1: 'AfricanAmerican',
    2: 'Asian',
    3: 'Other'
}
df['Ethnicity'] = df['Ethnicity'].map(ethnicity_map)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['Ethnicity'], prefix='Ethnicity', dtype=int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = df.drop(columns='Diagnosis'), df['Diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix


sns.heatmap(confusion_matrix(y_test, clf.predict(X_test)), annot=True)
```

כאן גיליתי שהמודל מפספס הרבה מקרים שהם באמת 1 והוא אומר שהם 0

```python
clf_v2 = RandomForestClassifier(max_depth=2, random_state=0, class_weight='balanced')
clf_v2.fit(X_train, y_train)
```

```python
clf_v2.score(X_train, y_train)
```

```python
clf_v2.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix


sns.heatmap(confusion_matrix(y_test, clf_v2.predict(X_test)), annot=True)
```
